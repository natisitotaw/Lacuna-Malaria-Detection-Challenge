```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Specify the extracted folder path
folder_path = 'data'  # Update with your actual path

# Get a list of all image files in the folder
image_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(('png', 'jpg', 'jpeg'))]

# Select 4 random images
random_images = random.sample(image_files, 4)

# Display the selected images
plt.figure(figsize=(10, 10))

for i, image in enumerate(random_images):
    img = mpimg.imread(image)
    plt.subplot(2, 2, i+1)
    plt.imshow(img)
    plt.axis('off')

plt.show()
```

```python
import os

def print_folder_tree(path, level=0):
    if not os.path.isdir(path):
        print(f"{path} is not a valid directory.")
        return

    prefix = "    " * level
    print(f"{prefix}|-- {os.path.basename(path)}/")

    try:
        for item in os.listdir(path):
            item_path = os.path.join(path, item)
            if os.path.isdir(item_path):  # Only consider directories
                print_folder_tree(item_path, level + 1)
    except PermissionError:
        print(f"{prefix}    |-- [Permission Denied]")

# Example usage
path = "/kaggle/working/data"  
print_folder_tree(path)
```
